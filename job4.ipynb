```python
#2)
def zad2(filename):
  with open(filename,'r',encoding= 'utf-8') as file:
    lines=file.readlines()
  toys =[]
  for line in lines:
    parts=line.strip().split(';')
    if len(parts)!= 3:
      continue
    if parts[0].lower == "мяч":
        continue
    min_age_diap, max_age_diap=map(int, parts[2].split('-'))
    if min_age_diap <= 3 <=max_age_diap:
      toys.append(parts[0])
  if toys:
    print("Подходящие игрушки кроме мяча для 3 лет : ")
    for toy in toys:
      print(toy)
  else:
    print("Нет подходящей игрушки")
zad2("Assort.txt")
```
