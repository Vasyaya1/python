```python
def delitel(x):
  k=0
  for i in range(1,int( x**0.5)+1):
    if x%i==0:
      k+=1
      if i!=x//i:
        k+=1
  return k
def rez(m, n):
  max_c=0
  num=[]
  for i in range(m, n+1):
    k=delitel(i)
    if max_c<k:
      max_c=k
      num = [i]
    elif k == max_c:
         num.append(i)
  return num

m=int(input())
n=int(input())
result= rez(m, n)
print(result)
```

```python
#2)
def Reduction(ch, zn):
  if zn==0 :
    print("ощщибка")
    return
  x=min(abs(ch), abs(zn))
  while x!=0:
    if ch%x==0 and zn%x==0:
      ch= ch/x
      zn=zn/x
    x=x-1
def sum(A,B,C,D):
  ch= A*D+B*C
  zn= B*D
  Reduction(ch,zn)
  print(ch,'/',zn)
a=int(input())
b=int(input())
c=int(input())
d=int(input())
sum(a,b,c,d)
```

```python
#3)
def last_str(Q):
  n =0
  total = 0
  while total+(n+1)<=Q:
    n=n+1
    total+=n
  return n
Q=int(input())
print(last_str(Q))
```
