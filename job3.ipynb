```python
#4)
def maxmin(x):
  max_x = min_x=x[0][0]
  max_x_i=max_x_j=min_x_i=min_x_j=0
  for i in range(len(x)):
    for j in range(len(x[i])):
      num = x[i][j]
      if num>max_x:
        max_x=num
        max_x_i=i
        max_x_j=j
      if num<min_x:
        min_x=num
        min_x_i=i
        min_x_j=j
  x[max_x_i][max_x_j], x[min_x_i][min_x_j]=x[min_x_i][min_x_j],x[max_x_i][max_x_j]
def print_mas(x):
  for i in range(len(x)):
     for j in range(len(x[i])):
       print(x[i][j],end=' ')
     print()

x=[
    [4, 8, 12],
    [66, 3, 15],
    [32, 2, 1]
]
maxmin(x)
print_mas(x)
```

```python
#1)
def frequency_of_words(sl:str):
  text=sl.split()
  dict1={}
  for word in text:
    clean_word =word.strip(".,!?;:").lower()
    if clean_word in dict1:
      dict1[clean_word]+=1
    else:
      dict1[clean_word]=1
  return dict1
sl=str(input())
res=frequency_of_words(sl)
print(res)
```

```python
#2)
def changing_places(dict1):
  dict2={}
  for key, val in dict1.items():
    if val in dict2:
      if isinstance(dict2[val], list):
        dict2[val].append(key)
      else:
        dict2[val]=[dict2[val], key]
    else:
      dict2[val] = key
  return dict2
dict1={1:'a', 2:'b', 3:'a'}
dict2=changing_places(dict1)
print(dict2)
```

```python
#3)
def merge(dist1, dist2):
  dist3=dist1.copy()
  for key in dist2:
    if key in dist3:
      dist3[key]+=dist2[key]
    else:
      dist3[key]=dist2[key]
  return dist3
dist1={'a':1, 'b':2, 'c':3}
dist2={'f':4, 'b':5, 'h':6}
dist3=merge(dist1, dist2)
print(dist3)
```

```python
#4)
def Limit(dist1, n):
  dist2={}
  for key,val in dist1.items():
    if isinstance(val,(int,float)) and val>n:
      dist2[key]=val
  return dist2
dist1={'a':1, 'b':2, 'c':3}
dist2=Limit(dist1, 2)
print(dist2)
```

```python
#5)
def Max(dist1):
  max_val=next(iter(dist1.values()))
  for key, val in dist1.items():
    if isinstance(val,(int,float)) and max_val<val:
      max_val=dist1[key]
  return max_val
dist1={'a':15, 'b':2, 'c':3}
a=Max(dist1)
print(a)
```

```python
#1)
def Unique(list1):
  set1=set()
  for i in range(0,len(list1)-1):
      if list1.count(i):
        set1.add(list1[i])
      else:
        continue
  return set1
list1 =[1,2, 1, 1,2]
set1=Unique(list1)
print(set1)
```

```python
#2)
def Del(list1):
  set1=set(list1)
  list2 = list(set1)
  return list2
list1 =[1,2, 1, 1,2, 6, 3, 6]
list2=Del(list1)
print(list2)
```

```python
#3)
def intersection(set1, set2):
  set3=set()
  for i in set1:
    if i in set2:
       set3.add(i)
  return set3
set1={10, 2 ,3, 8, 5, 5}
set2={3,5,7,2, 5, 4, 8}
set3=intersection(set1, set2)
print(set3)
```

```python
#4)
def unification(set1, set2):
  set3=set1.copy()
  for i in set2:
    set3.add(i)
  return set3
set1={10, 2 ,3}
set2={3,5,7}
set3=unification(set1, set2)
print(set3)
```

```python
#5)
def difference(set1, set2):
  set3=set()
  for i in set1:
      if i not in set2:
        set3.add(i)
  return set3
set1={10, 2 ,3, 8, 5, 5}
set2={3,5,7,2, 5, 4}
set3=difference(set1, set2)
print(set3)
```

```python
#6)
def symmetric_difference(set1, set2):
  diff1 = set1 - set2
  diff2 = set2 - set1
  return diff1.union(diff2)
set1={10, 2 ,3, 8, 5, 5}
set2={3,5,7,2, 5, 4}
set3=symmetric_difference(set1, set2)
print(set3)
```

```python
#7)
def subset(set1, set2):
  for i in set2:
    if i not in set1:
      return False
  return True
set1={1, 2, 3, 4}
set2={1, 3}
print(subset(set1, set2))
```

```python
#9)
def Push(set1, list1):
  for i in list1:
    set1.add(i)
set1={1, 3, 4}
list1=[2, 6, 7]
Push(set1, list1)
print(set1)
```

```python
#10)
def Pop(set1, list1):
  for i in list1:
    set1.discard(i)
set1={1, 3, 4}
list1=[1, 4]
Pop(set1, list1)
print(set1)
```

```python
#5)
def Count(t1, n):
  if n in t1:
    return t1.count(n)
  return -1
t1=(1, 3, 5, 1, 1, 3,2)
print(Count(t1, 3))
```

```python
#6)
def Max(t1):
  max_val=t1[0]
  for i in t1:
    if max_val<i:
      max_val=i
  return max_val
t1=(1, 3, 5, 1, 1, 3,2, 66)
print(Max(t1))
```

```python
#8)
def Sort(t1):
  list1=list(t1)
  n = len(list1)
  for i in range(n):
    for j in range(0, n-i-1):
      if list1[j]>list1[j+1]:
        list1[j],list1[j+1]=list1[j+1], list1[j]
  return list1
t1=(1, 3, 5, 1, 1, 3,2, 66)
print(Sort(t1))
```
