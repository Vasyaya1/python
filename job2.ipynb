```python
class Project:
  def __init__ (self, name, budget):
    self.name=name
    self.budget=budget
  def print_project_info(self):
    print(f"{self.name}, Бюджет: {self.budget}")
class Company:
  def __init__ (self, name_company):
    self.name_company=name_company
    self.projects=[]
  def add_project(self ,project: Project):
    self.projects.append(project)
  def print_company_info(self):
    print(f"Фирма {self.name_company}")
    print("список проектов: ")
    for pr in self.projects:
      pr.print_project_info()
project1=Project("Yaya", 12)
project2=Project("Tit", 88)
company=Company("Gasprom")
company.add_project(project1)
company.add_project(project2)
company.print_company_info()
```

```python
from abc import ABC, abstractmethod
class ListItem(ABC):
  @abc.abstractmethod
  def getName(self):
    pass
  def getPrice(self):
    pass
class BookItem(ListItem):
  def __init__ (self, name, price):
    self.name=name
    self.price=price
  def getName(self):
    return self.name
  def getPrice(self):
    return self.price
class ElectronicsItem(ListItem):
  def __init__ (self, name, price):
    self.name=name
    self.price=price
  def getName(self):
    return self.name
  def getPrice(self):
    return self.price
book=BookItem("Идиот", 1222)
elec=ElectronicsItem("Ноутбук", 50000)
print(f"Книга: {book.getName()}, Цена: {book.getPrice()} руб.")
print(f"Электроника: {elec.getName()}, Цена: {elec.getPrice()} руб.")
```
